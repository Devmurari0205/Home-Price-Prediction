##### Part-2: Working with Model 
##### 1) Create the target data and feature data where target data is price 


```python
X = df.drop('Price',axis=1)
y = df['Price']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split (X,y, test_size=30, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

###### 2) Create a linear regression model for Target and feature data 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

#### Linear Model

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
lr.fit(X_train_scaled,y_train)
```

```python
y_train_pred = lr.predict(X_train_scaled)
y_test_pred = lr.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
train_scor = lr.score(X_train_scaled,y_train)
test_scor = lr.score(X_test_scaled,y_test)

print("Training R2 Score :",train_scor)
print("Test R2 Score :",test_scor)
```

Mean Squared Error

```python
print("MSE Mean Squared error:",mean_squared_error(y_test,y_test_pred))
print("MAE Mean Absolute Error:",mean_absolute_error(y_test, y_test_pred))
print("RMSE Root Mean Squared Error",np.sqrt(mean_squared_error(y_test,y_test_pred)))
```

```python
r2 = r2_score(y_test, y_test_pred)
print("R2 Score:", r2)
```

##### If the model is overfitting then apply ridge and lasso regression algorithms 

##### alpha = Regularization strength; must be a positive float. Regularization
    improves the conditioning of the problem and reduces the variance of
    the estimates. Larger values specify stronger regularization.
max_iter : Maximum number of iterations for conjugate gradient solver.

```python
from sklearn.linear_model import Ridge
```

```python
ridge  = Ridge (alpha = 50,max_iter = 100) 
ridge.fit(X_train_scaled,y_train)
```

```python
ridge_pred = ridge.predict(X_test_scaled)
```

```python
print ("Ridge R2 Score"),r2_score(y_test,ridge_pred)
```

#### lasso Regresstion

```python
from sklearn.linear_model import Lasso
```

```python
lasso_reg =Lasso(alpha=50,max_iter=100)
lasso_reg.fit(X_train_scaled,y_train)
```

```python
lasso_pred = lasso_reg.predict(X_test_scaled)
```

```python
print("lasso R2 Score:",r2_score(y_test,lasso_pred))
```

```python
scores = {
    "Linear": r2_score(y_test, y_test_pred),
    "Ridge": r2_score(y_test, ridge_pred),
    "Lasso": r2_score(y_test, lasso_pred)
}
scores
```

```python
best_model = max(scores, key=scores.get)
print("Best Model:", best_model)
print("Best R2 Score:", scores[best_model])
```
